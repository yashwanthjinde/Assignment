```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# dates = pd.date_range('1/1/2000', periods=10)
df=pd.DataFrame(np.random.randint(45,100,size=(2000,3)), columns=['maths', 'scien', 'physics'])
df
array=np.array(df)
def total():
    total=[]
    sum1=0
    for i in range(0,2000):
        array[i]
        sum1=0
        for j in array[i]:
            sum1=sum1+j
        total.append(sum1)
    print(total)
    first=0
    sec=0
    third=0
    fail=0
    for k in total:
        if k>=270 and k<=300:
            first+=1
        elif k>=200 and k<=270:
            sec+=1
        elif k>=150 and k<=200:
            third+=1
        else:
            fail+=1
    print("first class",first)
    print("second class ",sec)
    print("third class",third)
    print("fail",fail)
    lang = ['first class', 'second class', 'fourth class', 'fail']
    rating = [first,sec,third,fail]
    plt.pie(rating, labels =lang,explode = [0,0,0.1,0], shadow = True , startangle = 90 ,autopct = '%1.1f%%', labeldistance = 1.2)
    plt.legend(loc = "best")

            
            
    
total()
        
```
