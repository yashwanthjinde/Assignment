```python
pcs=PCA()
pcs_df=pcs.fit_transform(data1)
```

```python
(pcs.components_)
pccomp=pd.DataFrame(pcs.components_)
pccomp
```

```python
var_df=pcs.explained_variance_ratio_
var_df
```

```python
comp=pd.DataFrame(pcs_df)
comp.head()
```

```python
cs=np.cumsum(np.round(var_df,decimals=4)*100)
cs
```

```python
plt.plot(cs,color='cyan')
plt.show()
```
